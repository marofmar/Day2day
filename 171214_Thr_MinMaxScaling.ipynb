단순히 파형만 본다면, Acceloremeter의 x, y, z 값을 min-max scaling 한 것이 의미가 있을 것이다.
그러나 상대적인 각 파형 사이의 간격이, 노쇠한 사람과 그렇지 않은 사람 사이에서 차이를 보이는 것으로 추측되는데,
이 경우에는 min-max scale 이 적절하지 않은 것이다. 안보이거든

우선 Accel 과 Gyro는 각 파형의 y축 값 차이가 유의미해 보이고 하니 Magna 로 한 번 봐야겠다

```python
# Min-Max Scaling

def min_max_scaling(df, col_list):
    cols = col_list
    df_scaled = df.copy()
    df_scaled[cols] = (df_scaled[cols]-df_scaled[cols].min() ) / (df_scaled[cols].max() - df_scaled[cols].min())
    return df_scaled
# returns the whole new data frame that certain colum values have been normalized
```

```python
def ref_min_max_scaling(df, col_list):
    cols = col_list
    df_scaled = df.copy()
    df_scaled[cols] = 2*{(df_scaled[cols]-df_scaled[cols].min() ) / (df_scaled[cols].max() - df_scaled[cols].min())}-1
    return df_scaled
# 논문에선 단순 민맥스 정규화에 2를 곱하고 1을 또 뺀다. 
```
