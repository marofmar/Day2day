```python
def get_acc(df): # 세 칼럼씩 꺼내기
    cut_df = df.iloc[:, 1:4]
    return cut_df
def get_gyr(df):
    cut_df = df.iloc[:, 4:7]
    return cut_df
def get_lat(df):
    cut_df = df.iloc[:, 7:]
    return cut_df
```
