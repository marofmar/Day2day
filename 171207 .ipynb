```python
def outlier_dist(df, col,name):
    plt.figure(figsize=(8,6))
    plt.scatter(range(df.shape[0]), np.sort(col.values))
    plt.xlabel('index', fontsize=12)
    plt.ylabel('name', fontsize=12)
    plt.show()
    
    #distribution
    ulimit = np.percentile(col.values, 99) # upper limit
    llimit = np.percentile(col.values, 1) # lower limit
    col.loc[col>ulimit] = ulimit
    col.loc[col<llimit] = llimit

    plt.figure(figsize=(12,8))
    sns.distplot(col.values, bins=50, kde=False)
    plt.xlabel('name', fontsize=12)
    plt.show()
```
